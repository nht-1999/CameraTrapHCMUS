```python
from pathlib import Path
import shutil
import os
```

```python
n = 9676
steps = int(n / (n * 0.1))
path = './yolo/datasets/'
```

```python
for i in range(n):
  imgPath = path + 'ena24/images/' + str(i+1) + '.jpg'
  labelPath = path + 'ena24/labels/' + str(i+1) + '.txt'

  if not Path(imgPath).is_file() or not Path(labelPath).is_file():
    print('error at: ', i+1)
  else:
    if i % steps == 0:
      shutil.copy(imgPath, path + 'ena24_splitted/test/images/' + str(i+1) + '.jpg')
      shutil.copy(labelPath, path + 'ena24_splitted/test/labels/' + str(i+1) + '.txt')
    else:
      shutil.copy(imgPath, path + 'ena24_splitted/train/images/' + str(i+1) + '.jpg')
      shutil.copy(labelPath, path + 'ena24_splitted/train/labels/' + str(i+1) + '.txt')
    
```
